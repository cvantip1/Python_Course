```python
shopping_cart = []
prices = {"apple": 1, "banana": 0.5, "milk": 2, "bread": 1.5}

def add_item():
    item = input("Enter item to add: ").lower()
    if item in prices:
        shopping_cart.append(item)
        print(f"{item.capitalize()} added to the cart.")
    else:
        print(f"Sorry, we don't have {item}.")

def remove_item():
    item = input("Enter item to remove: ").lower()
    if item in shopping_cart:
        shopping_cart.remove(item)
        print(f"{item.capitalize()} removed from the cart.")
    else:
        print(f"{item.capitalize()} is not in your cart.")

def view_cart():
    if shopping_cart:
        print("Your cart:")
        for item in shopping_cart:
            print(f" - {item.capitalize()} (${prices[item]})")
    else:
        print("Your cart is empty.")

def checkout():
    total = sum(prices[item] for item in shopping_cart)
    print(f"Your total is: ${total:.2f}")
    print("Thank you for shopping with us!")

def shopping():
    while True:
        print("\nWhat would you like to do?")
        print("1. Add item")
        print("2. Remove item")
        print("3. View cart")
        print("4. Checkout")
        print("5. Quit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            add_item()
        elif choice == "2":
            remove_item()
        elif choice == "3":
            view_cart()
        elif choice == "4":
            checkout()
            break
        elif choice == "5":
            break
        else:
            print("Invalid choice. Please try again.")

shopping()
```
