```python
# Dictionary to store students and their grades
students = {}

# Function to calculate average grade
def calculate_average():
    if students:
        average = sum(students.values()) / len(students)
        print(f"The average grade is: {average:.2f}")
    else:
        print("No students in the system yet.")

# Function to display highest and lowest grades
def display_highest_lowest():
    if students:
        highest = max(students.values())
        lowest = min(students.values())
        print(f"Highest grade: {highest}")
        print(f"Lowest grade: {lowest}")
    else:
        print("No students in the system yet.")

# Function to display passing or failing students
def show_passing_failing(passing=True):
    if students:
        threshold = 50
        if passing:
            print("Students passing:")
            for name, grade in students.items():
                if grade >= threshold:
                    print(f"- {name} ({grade})")
        else:
            print("Students failing:")
            for name, grade in students.items():
                if grade < threshold:
                    print(f"- {name} ({grade})")
    else:
        print("No students in the system yet.")

# Main function to manage the system
def classroom_grading_system():
    while True:
        print("\n1. Add a new student")
        print("2. Display average grade")
        print("3. Display highest grade")
        print("4. Display lowest grade")
        print("5. Show passing students")
        print("6. Show failing students")
        print("7. Quit")
        
        choice = input("Choose an option: ")

        if choice == "1":
            name = input("Enter student name: ")
            grade = int(input(f"Enter grade for {name}: "))
            students[name] = grade
            print(f"Student {name} with grade {grade} added.")
        
        elif choice == "2":
            calculate_average()
        
        elif choice == "3":
            display_highest_lowest()

        elif choice == "4":
            display_highest_lowest()

        elif choice == "5":
            show_passing_failing(passing=True)

        elif choice == "6":
            show_passing_failing(passing=False)
        
        elif choice == "7":
            print("Goodbye!")
            break
        
        else:
            print("Invalid option. Please choose again.")

# Run the grading system
classroom_grading_system()
```
